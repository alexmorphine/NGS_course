```python
result = pd.read_csv('quast_result/report.tsv', sep='\t').set_index('Assembly')
```

```python
# Genome fraction (%)
# сортировка от лучшего к худшему по показателю
# лучший - pacbio80, худший - pacbio10, разница достаточно большая
result.loc['Genome fraction (%)'].sort_values(ascending=False)
```

```python
# Duplication ratio
# сортировка от лучшего к худшему по показателю
# лучший - pacbio10, худший - pacbio20, разницы почти нет 
result.loc['Duplication ratio'].sort_values(ascending=True)
```

```python
# Largest alignment
# сортировка от лучшего к худшему по показателю
# лучший - pacbio80, худший - pacbio10, разница достаточно большая
result.loc['Largest alignment'].astype(int).sort_values(ascending=False)
```

```python
# Total aligned length
# сортировка от лучшего к худшему по показателю
# лучший - pacbio80, худший - pacbio10, разница лучшего и худшего достаточно большая, но в топе примерно одинаково
result.loc['Total aligned length'].astype(int).sort_values(ascending=False)
```

```python
# NGA50
# сортировка от лучшего к худшему по показателю
# лучший - pacbio80, худший - pacbio10, разница лучшего и худшего достаточно большая, но в топе примерно одинаково
result.loc['NGA50'].astype(int).sort_values(ascending=False)
```

```python
# LGA50
# сортировка от лучшего к худшему по показателю
# лучший - pacbio40, худший - pacbio10, разница лучшего и худшего достаточно большая, но в топе примерно одинаково
result.loc['LGA50'].astype(int).sort_values(ascending=True)
```

```python
# misassemblies
# сортировка от лучшего к худшему по показателю
# лучший - illumina100, худший - pacbio20, разница лучшего и худшего достаточно большая
result.loc['# misassemblies'].astype(int).sort_values(ascending=True)
```

```python
# Misassembled contigs length
# сортировка от лучшего к худшему по показателю
# лучший - illumina100, худший - pacbio80, разница лучшего и худшего достаточно большая
result.loc['Misassembled contigs length'].astype(int).sort_values(ascending=True)
```
